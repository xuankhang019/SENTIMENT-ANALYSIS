TEXT CLEANING AND PREPROCESSING

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('mebe_tiki.csv')
data
```

```python
import re
import string
import pickle
from spacy.lang.vi import Vietnamese
nlp = Vietnamese()
from spacy.lang.vi import STOP_WORDS
stop = STOP_WORDS
from nltk.tokenize.treebank import TreebankWordDetokenizer
from underthesea import word_tokenize
```

```python
def remove_xd(line):
    xdfree= re.sub(r'\n', ' ', line)
    return xdfree
```

```python
data['Comment']= data['Comment'].apply(lambda x:remove_xd(x))
data
```

```python
def remove_punctuation(line):
    punctuationfree= re.sub(r'[^\w\s]', ' ', line)
    return punctuationfree
```

```python
data['Comment']= data['Comment'].apply(lambda x: x.lower())
```

```python
data['Comment']= data['Comment'].apply(lambda x:remove_punctuation(x))
data
```

```python
def remove_emoji(line):
    emo_remov = re.sub('\W+',' ', line)
    return emo_remov
```

```python
data['Comment'] = data['Comment'].apply(lambda x : remove_emoji(x))
data
```

```python
def remove_numbers(line):
    rem_num = re.sub('[0-9]+', '', line)
    return rem_num
```

```python
data['Comment'] = data['Comment'].apply(lambda x : remove_numbers(x))
data
```

```python
def remove_stopwords(line):
    doc = nlp(str(line))
    filtered = [token.text for token in doc if token.is_stop == False]
    return filtered
```

```python
data['Comment'] = data['Comment'].apply(lambda x : remove_stopwords(x))
data
```

```python
def Detoken(line):
    afterdetoken=TreebankWordDetokenizer().detokenize(line)
    return afterdetoken
```

```python
data['Comment'] = data['Comment'].apply(lambda x : Detoken(x))
data
```

```python
def tokenize_two(line):
    tkn = word_tokenize(line, format="text")
    return tkn
```

```python
data['Comment'] = data['Comment'].apply(lambda x : tokenize_two(x))
data
```
